```python
import boto3
aws_access_key_id = ""
aws_secret_access_key = ""
```

```python
client = boto3.client(service_name='textract',region_name='us-east-2',aws_access_key_id=aws_access_key_id,aws_secret_access_key=aws_secret_access_key)
```

## 1. Start document analysis

```python
response = client.start_document_analysis(
        DocumentLocation={
        'S3Object': {
            'Bucket': 'textract-console-us-east-2-d9ab0c26-5378-4214-b4ff-c53c39ef7fc4',
            'Name': '62739298322_2021-10-16.pdf',
        }
    },
    FeatureTypes=[
        'TABLES',
    ],
)

print(response)
```

## 2. Get document analysis

```python
analysis = client.get_document_analysis(
    JobId='5b4733c97ae6bac26980c41538fffb791a7712a328c3feb0fa143311a0ac3b2a',
)
```
